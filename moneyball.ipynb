```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import re

%matplotlib inline
```

OBP = (H + BB + HBP)/(AB + BB + HBP + SF)

```python
salaries = pd.read_csv('data/salaries.csv')
```

```python
batting = pd.read_csv('data/batting.csv')
```

```python
fielding = pd.read_csv('data/fielding.csv')
```

```python
master = pd.read_csv('data/master.csv')
```

```python
salaries.head()
```

```python
batting.head()
```

```python
stats = batting[['playerID', 'yearID','H', 'BB', 'HBP', 'AB', 'SF']]
```

```python
stats.head()
```

```python
fielding.head()
```

```python
position = fielding[['playerID', 'yearID', 'POS', 'G']]
```

```python
position.head()
```

```python
name_data = master[['playerID', 'nameFirst', 'nameLast']]
```

```python
name_data.head()
```

```python
all_stats = pd.merge(position, stats)
```

```python
all_stats.head()
```

```python
all_stats_salary = pd.merge(all_stats, salaries)
```

```python
all_stats_salary.head()
```

```python
stats_salary_name = pd.merge(name_data, all_stats_salary)
```

```python
stats_salary_name.head()
```

```python
stats_salary_name['OBP'] = (stats_salary_name['H']+stats_salary_name['BB']+stats_salary_name['HBP'])/(stats_salary_name['AB']+stats_salary_name['BB']+stats_salary_name['HBP']+stats_salary_name['SF'])
```

```python
stats_salary_name.head()
```

```python
stats_salary_name.sort('OBP', ascending=False)
```

```python
twenty_ab = stats_salary_name['AB'] >= 25
```

```python
twenty_ab = stats_salary_name[twenty_ab]
```

```python
twenty_ab.sort('OBP', ascending=False).head()
```

```python
all_data = twenty_ab.copy()
```

```python
all_data['OBP/salary'] = twenty_ab['OBP']/twenty_ab['salary']
```

```python
all_data.sort(['OBP/salary'], ascending=False)
```

```python
final_data = all_data[['playerID', 'nameFirst','nameLast', 'yearID', 'POS', 'salary', 'OBP', 'OBP/salary']]
```

```python
final_data = final_data.sort(['OBP/salary'], ascending=False)
```

```python
final_data
```
